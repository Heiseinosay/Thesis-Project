```python
# human_file = 'AudioData/human/spanish86.mp3'
debussy_file = 'audio/debussy.wav'
ipd.Audio(human_file)
```

```python
audio, sr = librosa.load(debussy_file)
```

```python
print(type(audio))
```

```python
FRAME_LENGTH = 1024
HOP_LENGTH = 512
sr = 22050
```

# Spectral Centroid

```python
sc_audio = librosa.feature.spectral_centroid(y=audio, sr=sr, n_fft=FRAME_LENGTH, hop_length=HOP_LENGTH)[0]
```

```python
sc_audio.shape
```

```python
frames = range(len(sc_audio))
t = librosa.frames_to_time(frames)

plt.figure(figsize=(25, 10))
plt.plot(t, sc_audio, color="b")
plt.show()
```

# Spectral Bandwidth

```python
sb_audio = librosa.feature.spectral_bandwidth(y=audio, sr=sr, n_fft=FRAME_LENGTH, hop_length=HOP_LENGTH)[0]
```

```python
sb_audio.shape
```

```python
plt.figure(figsize=(25, 10))

plt.plot(t, sc_audio, color="b")
plt.plot(t, sb_audio, color="r")
plt.show()
```
