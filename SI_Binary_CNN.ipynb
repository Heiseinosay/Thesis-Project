```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import joblib
```

```python
"""
!!! NOTE : Need to replace the speaker_profile
"""

# Load the data
# FULL
speaker_profile = pd.read_csv('result\\voice_recognition\\training_full\\vr_speaker_segment.csv') 
others_profile = pd.read_csv('result\\voice_recognition\\training_full\\vr_other_segment.csv')

# MEAN
# speaker_profile = pd.read_csv('result\\voice_recognition\\training_mean\\vr_speaker_mean_features2.csv')
# others_profile = pd.read_csv('result\\voice_recognition\\training_mean\\vr_other_mean_features2.csv')
```

```python
speaker_profile
```

```python
speaker_profile['label'] = 1
others_profile['label'] = 0
```

```python
data_df = pd.concat([speaker_profile, others_profile], ignore_index=True)
# Shuffle the data
data_df = data_df.sample(frac=1).reset_index(drop=True)
```

```python
# Split features and labels
X = data_df.drop(columns=['file_name', 'label']).values
y = data_df['label'].values
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.01, random_state=42)
```

```python
# Standardize the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Save the scaler
# joblib.dump(scaler, 'vr_scaler.pkl')
```

```python
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1, 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1, 1))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Input

# # Define the model
# speaker_identification_model = Sequential([
#     # Convolutional Layers
#     Conv2D(32, (3, 3), activation='relu', input_shape=(X_train.shape[1], 1, 1), padding='same'),
#     # Pooling Layers
#     MaxPooling2D((2, 1)),
#     Conv2D(64, (3, 3), activation='relu', padding='same'),
#     MaxPooling2D((2, 1)),
#     Conv2D(128, (3, 3), activation='relu', padding='same'),
#     MaxPooling2D((2, 1)),
#     Conv2D(256, (3, 3), activation='relu', padding='same'),
#     MaxPooling2D((2, 1)),
#     # Flatten Layer: Convert the feature maps into a 1D vector.
#     Flatten(),
#     # Hidden Layers: Extract and transform features from the flattened vector.
#     Dense(256, activation='relu'),
#     Dropout(0.5),
#     Dense(128, activation='relu'),
#     Dropout(0.5),
#     # Output Layer: Produces the final prediction. activation function = sigmoid for binary classification
#     Dense(1, activation='sigmoid')
# ])

# # Compile
# # Optimizer to consider 'SGD' | RMSprop 
# speaker_identification_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# speaker_identification_model.summary()

# Second method - reducing the complexity of the model
speaker_identification_model = Sequential([
    # Convolutional Layers
    Conv2D(16, (3, 3), activation='relu', input_shape=(X_train.shape[1], 1, 1), padding='same'),
    MaxPooling2D((2, 1)),  # Pooling Layer
    
    Conv2D(32, (3, 3), activation='relu', padding='same'),
    MaxPooling2D((2, 1)),  # Pooling Layer
    
    # Flatten Layer: Convert the feature maps into a 1D vector
    Flatten(),
    
    # Hidden Layer: Extract and transform features from the flattened vector
    Dense(64, activation='relu'),
    Dropout(0.5),
    
    # Output Layer: Produces the final prediction for binary classification (sigmoid activation)
    Dense(1, activation='sigmoid')
])


# Compile
# Optimizer to consider 'SGD' | RMSprop 
speaker_identification_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

speaker_identification_model.summary()
```

```python
# Train the model
history = speaker_identification_model.fit(X_train, y_train, epochs=20, batch_size=16, validation_split=0.2)
```

```python
# Evaluate the model on the test set
test_loss, test_acc = speaker_identification_model.evaluate(X_test, y_test)
print(f'Test Accuracy: {test_acc:.4f}')
```
