# Other accents

```python
df.accent.unique()
```

```python
accent_us = df[df['accent'] == 'us'].head(50)
accent_england = df[df['accent'] == 'england'].head(50)
accent_indian = df[df['accent'] == 'indian'].head(50)
accent_canada = df[df['accent'] == 'canada'].head(50)
accent_malaysia = df[df['accent'] == 'malaysia'].head(50)
accent_ireland = df[df['accent'] == 'ireland'].head(50)
accent_bermuda = df[df['accent'] == 'bermuda'].head(50)
accent_scotland = df[df['accent'] == 'scotland'].head(50)
accent_african = df[df['accent'] == 'african'].head(50)
accent_newzealand = df[df['accent'] == 'newzealand'].head(50)
accent_wales = df[df['accent'] == 'wales'].head(50)
accent_singapore = df[df['accent'] == 'singapore'].head(50)
accent_hongkong = df[df['accent'] == 'hongkong'].head(50)
accent_southatlandtic= df[df['accent'] == 'southatlandtic'].head(50)
```

```python
human_df = pd.concat([accent_us, accent_england, accent_indian, accent_canada, accent_malaysia, accent_ireland, accent_bermuda, accent_scotland, accent_african, accent_newzealand, accent_wales, accent_singapore, accent_hongkong, accent_southatlandtic])
```
